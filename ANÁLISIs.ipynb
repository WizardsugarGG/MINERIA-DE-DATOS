```python
import pandas as pd

import matplotlib.pyplot as plt

df=pd.read_csv('streaming_platform.csv')
df.info()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="age",y="hours_watched_per_week")
plt.title("Edad vs horas vistas por semana")
plt.xlabel("Edades")
plt.ylabel("Horas vista por semana")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="hours_watched_per_week",y="account_creation_year")
plt.title("horas vista por semana vs Año de creacion de cuenta")
plt.xlabel("Horas vista por semana")
plt.ylabel("Año de creacion de cuenta")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="number_of_sessions",y="days_since_last_login")
plt.title("Numero de sesiones vs Ultimo dia de sesion")
plt.xlabel("Numero de sesiones")
plt.ylabel("Ultimo dia de sesion")
plt.show()
```
