```python
import pandas as pd
import os
print(os.getcwd())
# Leer archivos
orders_a = pd.read_csv("part2_orders_a.csv")
orders_b = pd.read_csv("part2_orders_b.csv")

# Unir en un solo DataFrame y reiniciar índice
orders = pd.concat([orders_a, orders_b], ignore_index=True)

print(orders.head())
```

```python
# Convertir a datetime por seguridad
orders["order_date"] = pd.to_datetime(orders["order_date"])

# Ordenar
orders = orders.sort_values(
    by=["order_date", "qty"],
    ascending=[True, False]
)

print(orders.head())
```

```python
orders = orders.set_index("order_date").sort_index()

print(orders.head())
```

Merge con clientes (many_to_one)

```python
# Leer clientes
print(os.listdir())
customers = pd.read_csv("customers.csv")

# Merge tipo left y validar many_to_one
orders = orders.merge(
    customers[["customer_id", "customer_name", "city", "segment"]],
    on="customer_id",
    how="left",
    validate="many_to_one"
)

print(orders.head())
```
