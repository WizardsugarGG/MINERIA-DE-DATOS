4. Análisis de Correlaciones y Selección
Para medir la relación, usamos el coeficiente de Pearson (de -1 a 1).

Variables altamente correlacionadas: Generalmente, Qty y Amount presentan la correlación más alta, ya que a mayor cantidad de productos, mayor es el monto total.

¿Por qué es útil una alta correlación?: Significa que existe una relación lineal fuerte. Si una variable sube y la otra también, el modelo puede usar la variable independiente para predecir con precisión el comportamiento del target.
